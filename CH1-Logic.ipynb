```python
def implies(p , q):
    return (not p or q)

assert implies(False, False) == True
assert implies(False, True)  == True
assert implies(True, False)  == False
assert implies(True, True)   == True
print("All Tests passed! Alhamdullillah!")
```

```python
def xor(p, q):
    return p != q
assert xor(False, False) == False
assert xor(False, True)  == True
assert xor(True, False)  == True
assert xor(True, True)   == False
print("All Tests passed! Alhamdullillah!")
```

```python
def iff(p, q):
    return p == q
assert iff(False, False) == True
assert iff(False, True)  == False
assert iff(True, False)  == False
assert iff(True, True)   == True
print("All Tests passed! Alhamdullillah!")
```

```python
# ∼p ∨ q → ∼q
for p, q in [ (False, False), (False, True), (True, False), (True, True) ]:
    e = implies((not p or q), not q)
    print(p , q, e)
```

```python
# (p →r ) ↔ (q →r )
for p, q, r in [ 
(False, False, False), 
(False, False, True), 
(False, True, False),
(False, True, True),
(True, False, False),
(True, False, True),
(True, True, False),
(True, True, True)
]:
    e1 = implies(p, r)
    e2 = implies(q, r)
    e = iff(e1, e2)
    print(p , q, r, e)
```

```python
def bitwise_and(bs1, bs2):
    result = []
    for i, j in zip(bs1, bs2):
            r = i and j
            result.append(r)
    return "".join(result)

def bitwise_or(bs1, bs2):
    result = []
    for i, j in zip(bs1, bs2):
            r = i or j
            result.append(r)
    return "".join(result)

def bitwise_xor(bs1, bs2):
    result = []
    for i, j in zip(bs1, bs2):
            r = xor(i, j)
            result.append(r)
    return "".join(result)

bitwise_and("110", "010")
```

```python
def bitwise(bs1, bs2, op):
    assert len(bs1) == len(bs2)
    if op == 'and':
        return bitwise_and(bs1, bs2)
    elif op == 'or':
        return bitwise_or(bs1, bs2)
    elif op == 'xor':
        return bitwise_xor(bs1, bs2)
    else:
        print("Sorry, The OP still under development!")

bitwise("11100010101010", "10100110101000", 'and')
```

## Programming Problem 1: Checking Consistency of System Specifications

### Problem
Write a Python program that checks whether a system of propositional statements is consistent.

A system is **consistent** if there exists at least one assignment of truth values that makes **all** statements true.

Use the following propositions:

- `M`: the system is in multiuser state
- `N`: the system is operating normally
- `K`: the kernel is functioning
- `I`: the system is in interrupt mode

Your program should test all possible truth assignments and determine whether the specifications are consistent.

### Specifications
Translate these statements into logic:

1) The system is in multiuser state if and only if it is operating normally.  
2) If the system is operating normally, then the kernel is functioning.  
3) The kernel is not functioning, or the system is in interrupt mode.  
4) If the system is not in multiuser state, then the system is in interrupt mode.  
5) The system is not in interrupt mode.  

### Goal
Your program should:

1) Generate all possible truth assignments for `M`, `N`, `K`, and `I`.  
2) Test whether all five specifications are true.  
3) Report whether the system is consistent.

## Programming Problem 2: Logical Equivalence of Two Expressions

### Problem
Write a Python function `equivalent(expr1, expr2, var_names)` that determines whether two propositional expressions are logically equivalent.

Two expressions are **logically equivalent** if they have the same truth value for every possible assignment of truth values to their variables.

### Input
Your function should take:

1) `expr1`: the first logical expression as a Python function  
2) `expr2`: the second logical expression as a Python function  
3) `var_names`: a list of variable names used in the expressions  

### Goal
Your function should:

1) Generate all possible truth assignments for the variables in `var_names`.  
2) Evaluate both expressions on each assignment.  
3) Return `True` if the two expressions always match.  
4) Return `False` otherwise.
